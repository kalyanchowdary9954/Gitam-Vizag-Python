- this used for math,scientific,data science programming

```python
import numpy as np 
li = [1,2,3,4,5,6]
c1 = np.array(li)
print(c1)
print(type(c1))
```

## visualization data
- an intresting way to display dataframe
- with matplotilb

```python
import matplotlib.pyplot as plt
a1 = [1,2,3,4]
a2 = [5,6,7,8]
plt.figure(figsize=(4,4))
plt.plot(a1,a2,color='red')
plt.xlabel('x-Axis')
plt.ylabel('Y-Axis')
plt.show
```

```python
marks = [79,84,75,69]
divisions = ['A','B','C','D']
index = np.arange(len(divisions))
plt.bar(divisions,marks,color='green',width=0.4)
```

```python
plt.barh(divisions,marks,color='black')
plt.show()
```

```python
lang = ['python','java','c','c++']
rating = [95,52,32,48]
plt.pie(rating,labels = lang,shadow = True,startangle = 90,autopct = '%1.1f%%')
plt.legend(loc='best')
plt.show()
```
